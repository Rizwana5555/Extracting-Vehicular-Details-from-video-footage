```
import os

# Create the directory if it doesn't exist
os.makedirs("/content/Traffic-Vehicle-Detection1", exist_ok=True)

# Now write the yaml file
yaml_text = """
path: /content/drive/MyDrive/Traffic Dataset
train: /content/drive/MyDrive/Traffic Dataset/images/train
val: /content/drive/MyDrive/Traffic Dataset/images/val

names:
  - Car
  - Number Plate
  - Blur Number Plate
  - Two Wheeler
  - Auto
  - Bus
  - Truck
"""

with open("/content/Traffic-Vehicle-Detection1/data.yaml", "w") as f:
    f.write(yaml_text)
```

```
from ultralytics import YOLO

model = YOLO("yolov12s.pt")

model.train(
    data="/content/Traffic-Vehicle-Detection1/data.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    name="vehicle detection",
    augment=True
)
```
